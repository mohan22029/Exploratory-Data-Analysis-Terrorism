```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("globalterrorismdb_0718dist.csv", encoding="latin1")
df = pd.DataFrame(data)
print("Data has been successfully imported")
df.head()
```

```python
def clean_data(df):
    df = df[["iyear", "country_txt", "attacktype1_txt", "weaptype1_txt", "nkill", "nwound", "region_txt"]]
    df.rename(columns={"iyear": "Year", "country_txt": "Country",
                       "attacktype1_txt": "Attack Type", "weaptype1_txt": "Weapon Type",
                       "nkill": "Killed", "nwound": "Wounded"}, inplace=True)
    df["Killed"] = df["Killed"].fillna(0)
    df["Wounded"] = df["Wounded"].fillna(0)
    return df
```

```python
def plot_bar(data, x, y, title, xlabel, ylabel):
    data.plot(kind="bar", color="cornflowerblue", figsize=(15, 6))
    plt.title(title, fontsize=15)
    plt.xlabel(xlabel, fontsize=15)
    plt.ylabel(ylabel, fontsize=15)
    plt.show()
```

```python
df = clean_data(data)
```

```python
attacks = df["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index": "Year", "Year": "Attacks"}).set_index("Year")
plot_bar(attacks, "Years", "Number of Attacks", "Timeline of Attacks", "Years", "Number of Attacks")
```

```python
yc = df[["Year", "Killed", "Wounded"]].groupby("Year").sum()
plot_bar(yc, "Years", "Number of Casualties", "Year-wise Casualties", "Years", "Number of Casualties")
```

```python
reg = pd.crosstab(df.Year, df.region_txt)
reg.plot(kind="area", stacked=False, alpha=0.5, figsize=(20, 10))
plt.title("Region-wise Attacks", fontsize=20)
plt.xlabel("Years", fontsize=20)
plt.ylabel("Number of Attacks", fontsize=20)
plt.show()
```

```python
regt = reg.transpose()
regt["Total"] = regt.sum(axis=1)
ra = regt["Total"].sort_values(ascending=False)
plot_bar(ra, "Region", "Number of Attacks", "Total Number of Attacks in Each Region", "Region", "Number of Attacks")
```

```python
rw = df[["region_txt", "Killed", "Wounded"]].groupby("region_txt").sum().sort_values(by="Wounded", ascending=False)

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(15, 6))
plot_bar(rw["Killed"], "Regions", "Number of People Killed", "People Killed in Each Region", "Regions", "Number of People Killed")
plot_bar(rw["Wounded"], "Regions", "Number of People Wounded", "People Wounded in Each Region", "Regions", "Number of People Wounded")

plt.show()
```
